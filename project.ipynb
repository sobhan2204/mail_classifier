```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split  # used to convert numerical value into test case 
from sklearn.feature_extraction.text import TfidfVectorizer  #used to convert mail data into numerical value
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import csv
```

```python
# DATA COLLECTION AND PRE PROCESSING 
data = pd.read_csv("C:/Users/SOBHAN/OneDrive/Documents/mail_data - mail_data.csv")
data.columns
```

```python
#replace null value with null string
mail_data = data.where((pd.notnull(data)),'') # IT WILL REPLACE ALL THE NULL VALUE WITH EMPTY SPACE 
data.head(5)
```

```python
print("The total numebr of elements in dataset is " , mail_data.size)
print("The total number of rows and columns in dataset is " , mail_data.shape)
```

```python
# WE WILL REPLACE ALL THE HAM VALUE TO 1
# AND ALL THE SPAM VALUE TO 0
mail_data['Category'] = mail_data['Category'].str.replace("ham",'1')
mail_data['Category'] = mail_data['Category'].str.replace("spam","0")
mail_data.head()
```

```python
# SEPERATING DATA AS TEXT AND LABLE 
x = mail_data['Message']
y = mail_data['Category']
print(x)
print(y)
```

```python
 # SPLITTING THE DATA INTO TRAINING AND TESTING DATA
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=3)#test_size is used to decide how much data u want for testing(i.e 0.2 means 20% of total data)
# if you want to have distributed data then you can mention random_state as 3 
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x.shape)#all data
print(x_test.shape)# 20% of data
print(x_train.shape)# 80% of dataA
```

```python
# FEATURE EXTRATION
 # TRANSFORM THE TEXT DATA INTO FEATURE VECTOR THAT CAN BE USED AS INPUT TO THE LOGISTIC REGRESSION
feature_extration = TfidfVectorizer(min_df=1 , stop_words='english' , lowercase=True) #tfidfvector goes through all the 
#words in your text and see how many time it is reaped or some other factor. Then it gives score accordingly and relate it 
# with 0 and 1 as hamor spam. This is how logesctic regression is trained
# min_df = if the score is lower then 1 or 1 ignore it
# stop_words = ignore the common words like the, a, an etc
# lowercase = convert all the text to lower case
```

```python
x_train_features = feature_extration.fit_transform(x_train) # used to fit the data and train it 
x_test_features = feature_extration.transform(x_test) # used to transform the data and we don't fit any data anymore
y_train
print(x_train_features) # all the text   is converted into numerical value
```

```python
# LOGESTIC REGRESSION
model1 = LogisticRegression()
# TRAINING THE REGRESSION MODEL WITH TRAINING DATA
model1.fit(x_train_features,y_train) # we don't need to transform y because it is already in int format
```

```python
# EVALUATING OUR MODEL ON TRAINIGN DATA
pridiction_on_training_model = model1.predict(x_train_features) # 
accuracy_on_training_model = accuracy_score(y_train,pridiction_on_training_model) # Tells how much accurate the model is 
print("Accuracy score of model: ", accuracy_on_training_model)
print("Accuracy on training model: ", accuracy_on_training_model*100,"%")
# A model with acurace rate of over of 75% is a good model 
```

```python
# EVALUATING OUR MODEL ON TEST DATA TO OVERCOME OVERFITTING 
pridiction_on_test_model = model1.predict(x_test_features) # 
accuracy_on_test_model = accuracy_score(y_test,pridiction_on_test_model) # Tells how much accurate the model is 
print("Accuracy score of model: ", accuracy_on_test_model)
print("Accuracy on training model: ", accuracy_on_test_model*100,"%")
```

```python
# BULDING A PREDECTIVE SYSTEM 
input_main = ["Go until jurong point, crazy.. Available only in bugis n great world la e buffet... Cine there got amore wat..."]
 # CONVERT ALL THE TEXT OF MAIL INTO FECTURE VECTOR 
input_data_feature  = feature_extration.transform(input_main)

# MAKE PREDECTION 
pridiction = model1.predict(input_data_feature)
print(pridiction)

# IF ELSE CONDITION
if pridiction[0] == 0:
    print("The mail is spam")
else:
    print("The mail is ham")
```

```python
# MAKING USER INTERACTIVE 
a = input("Enter the mail you want to find that it is spam or ham : ")
input_main = [a]
input_data_feature  = feature_extration.transform(input_main)

# MAKE PREDECTION 
pridiction = model1.predict(input_data_feature)
print(pridiction)
# OPEN CSV FILE TO STORE RESULT
with open('mail_classification_results.csv', mode='a', newline='') as file:
    writer = csv.writer(file)
    if pridiction[0] == 0:
        result = "spam"
        writer.writerow([a, result])  # Store the input and result
    else:
        result = "ham"
        writer.writerow([a, result])  # Store the input and result

print(f"The mail is {result}")  # Output is inside the 'with' block, no need to access file afterward
```
